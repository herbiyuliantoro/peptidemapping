```python
# Import required libraries
import pandas as pd

# Load the data
file_path = r"D:\DirectInfusion\Anti-body Monoclonal\replication of DI 1\1_filter.txt"# Replace with the path to your file
data = pd.read_csv(file_path, sep='\t')

# Step 1: Sort the data based on peptide names (alphabetically) and cosine scores (descending)
data_sorted = data.sort_values(by=['Peptide', 'cosine'], ascending=[True, False])

# Step 2: Filter the data
# Keep the highest cosine similarity score for each peptide name and retain rows with different charges
filtered_data = data_sorted.drop_duplicates(subset=['Peptide', 'z'], keep='first')

# Step 3: Save the filtered data to a new file or display it
filtered_file_path = r"D:\DirectInfusion\Anti-body Monoclonal\replication of DI 1\1_filter.csv"# Replace with your desired output file name
filtered_data.to_csv(filtered_file_path, index=False)

# Display the filtered data for confirmation
filtered_data
```
